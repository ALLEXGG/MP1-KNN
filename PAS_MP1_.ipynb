```python
import numpy as np
import pandas as pd

# Assuming 'sensus_df' is the DataFrame with your lung cancer data
# Check the actual column names in your DataFrame using sensus_df.columns
# and update the code below accordingly

# Example: If your columns are named 'age' and 'smoking', change the code to:
# x_train = np.array(sensus_df[['age', 'smoking']])
# y_train = np.array(sensus_df['lung_cancer'])

# If the column names are indeed 'AGE' and 'SMOKING', the issue might be with
# the DataFrame itself or how it was loaded. Double-check the data loading process.

# For demonstration purposes, assuming the column names are 'age', 'smoking', and 'lung_cancer':
x_train = np.array(df[['AGE', 'SMOKING']])  # Replace with actual column names
y_train = np.array(df['LUNG_CANCER'])  # Replace with actual column name

print(f'x_train:\n{x_train}\n')
print(f'y_train: {y_train}')
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(f'y_train:\n{y_train}')
```

```python
y_train = y_train.flatten()
print(f'y_train: {y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(x_train, y_train)
```

```python
Umur = 69
Perokok = 1
X_new = np.array([Umur, Perokok]).reshape(1,-1)
X_new
```

```python
y_new = model.predict(X_new)
y_new
```

```python
lb.inverse_transform(y_new)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for jk, d in df.groupby('LUNG_CANCER'):
    ax.scatter(x=d['AGE'], y=d['SMOKING'], label=jk)

plt.scatter(Umur,
            Perokok,
            marker='s',
            color='red',
            label='misterius')

plt.legend(loc='upper left')
plt.title('Sebarkan Data Umur, Yang merokok, dan Yang terkena kanker paru-paru')
plt.xlabel('Umur')
plt.ylabel('Perokok')
plt.grid(True)
plt.show()
```

```python
from scipy.spatial.distance import euclidean

data_jarak = [euclidean(misterius, d) for d in x_train]
data_jarak
```

```python
df['jarak'] = data_jarak
df.sort_values(['jarak'])
```

```python
x_test = np.array([[69, 1], [44, 1], [21, 2], [38, 1]])
y_test = lb.transform(np.array(['YES', 'YES', 'NO', 'YES'])).flatten()
print(f'x_test:\n{x_test}\n')
print(f'y_test:\n{y_test}')
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)

print(f'Accuracy: {acc}')
```

```python
from sklearn.metrics import precision_score

prec = precision_score(y_test, y_pred)

print(f'Precision: {prec}')
```

```python
from sklearn.metrics import recall_score

rec = recall_score(y_test, y_pred)

print(f'Recall: {rec}')
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print(f'F1 Score: {f1}')
```

```python
from sklearn.metrics import classification_report

cls_report = classification_report(y_test, y_pred)

print(f'Classification Report:\n{cls_report}')
```

```python
from sklearn.metrics import matthews_corrcoef

mcc = matthews_corrcoef(y_test, y_pred)

print(f'MCC:{mcc}')
```
